```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import folium
import seaborn as sns
```

```python
df = pd.read_csv('diamonds.csv')
df.head()
```

```python
df.info()
```

```python
plt.figure(figsize=(15,10))
sns.set(font_scale=2)
#sns.color_palette("magma")
sns.scatterplot(x='carat',y='price',data=df, color='purple',)
```

```python
plt.figure(figsize=(12,8))
sns.distplot(df['price'],kde=False, color='red')
plt.xlim(0,18000)
```

```python
sns.get_dataset_names()
```

```python
#tips = pd.read_csv('tips')
tips = sns.load_dataset('tips')
tips.head()
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(data=tips, x="total_bill", y="tip", color='red')
```

```python
plt.figure(figsize=(12,8))
sns.countplot(x='cut',data=df)
```

```python

plt.figure(figsize=(12,8))
sns.boxplot(x='cut',y='price',data=df)
```

```python
df['cut'].unique()
```

```python
cut_order = ['Fair','Ideal','Good', 'Very Good','Premium' ]
plt.figure(figsize=(12,8))
sns.boxplot(x='cut',y='price',data=df,order=cut_order,palette='cool')
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(14,10))
sns.set_context('paper', font_scale=1.4)

sns.jointplot(x='price', y='carat', data=df, kind='reg')
```

```python


sns.barplot(x='price', y='carat', data=df,estimator=np.median)
```

```python
with sns.axes_style('white'):
    sns.jointplot("carat", "price", data=df, kind='hex')
```

```python
plt.figure(figsize=(8,6))

sns.set_style('dark')

sns.set_context('talk')

sns.stripplot(x='depth',y='table',data=df, hue='color', 
             palette='seismic')
plt.legend(loc=0)
```
